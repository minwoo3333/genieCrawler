```python
# 1) requests 라이브러리 import
import requests

# 2) headers란 변수에 User-Agent 값 넣기
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'}

# 3) URL 리스트 설정
urls = [
    'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240521&hh=01&rtm=Y&pg=1',
    'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240521&hh=01&rtm=Y&pg=2',
    'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240521&hh=01&rtm=Y&pg=3',
    'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240521&hh=01&rtm=Y&pg=4'
]

# 첫 번째 URL에 대해 get 방식으로 HTML 문서 요청
data = requests.get(urls[0], headers=headers)
print(f'HTTP 상태 코드: {data.status_code}')  # HTTP 응답 상태 코드 확인
print(data.text[:1000])  # 응답 데이터의 일부 출력 (앞부분 1000자)
```

```python
from bs4 import BeautifulSoup

# 2) data에서 받은 HTML 문서를 파싱하여 soup 객체 생성
soup = BeautifulSoup(data.text, 'html.parser')

# trs 라는 변수에 아래 경로에 있는 모든 tr 태그 안에 있는 정보 가져오기
trs = soup.select('#body-content > div.newest-list > div > table > tbody > tr')
print(f'추출된 tr 태그 수: {len(trs)}')  # 추출된 tr 태그 수 확인
```

```python
# trs[0]부터 순서대로 돌면서 순위, 제목, 아티스트 정보 추출
for tr in trs:
    rank = tr.select_one('td.number').text.strip().split()[0]  # 순위는 첫 번째 단어만 추출
    title = tr.select_one('a.title.ellipsis').text.strip()
    artist = tr.select_one('a.artist.ellipsis').text.strip()

    # rank, title, artist 출력
    print(rank, title, artist)
```
